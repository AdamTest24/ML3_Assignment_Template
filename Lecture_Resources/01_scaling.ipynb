# Normalisation

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer 
from sklearn.preprocessing import MinMaxScaler

from numpy import amax, array
from numpy.random import randint

from matplotlib.pyplot import subplots
```

```python
my_data = randint(10, size=(5, 2))

my_data
```

## Standard Scaler

Standardize features by removing the mean and scaling to unit variance


```python
std_skl          = StandardScaler()
some_data_normed = std_skl.fit_transform(some_data)

print('Normalised data:', '\n', some_data_normed)
print('')

fig, ax = subplots(nrows=1, ncols=2)

scaling = amax(some_data)*1.1

ax[0].scatter(some_data[:, 0], some_data[:, 1])
ax[0].set_xlabel('Some data')

ax[1].scatter(some_data_normed[:, 0], some_data_normed[:, 1], c='r')
ax[1].set_xlabel('Normalised data');
```

```python
std_skl.transform([[2.3, 1.7]])
```

## Normaliser

Scale input vectors individually to unit norm (vector length).

```python
norm_skl         = Normalizer()
some_data_normed = norm_skl.fit_transform(some_data)

print('Normalised data:', '\n', some_data_normed)
print('')


fig, ax = subplots(nrows=1, ncols=2)

scaling = amax(some_data)*1.1

ax[0].scatter(some_data[:, 0], some_data[:, 1])
ax[0].set_xlim(0, scaling)
ax[0].set_ylim(0, scaling)
ax[0].set_xlabel('Some data')

ax[1].scatter(some_data_normed[:, 0], some_data_normed[:, 1], c='r')
ax[1].set_xlim(0, scaling)
ax[1].set_ylim(0, scaling);
ax[1].set_xlabel('Normalised data');
```

```python
norm_skl.transform([[2.3, 1.7]])
```

## Min-Max-Scaler

Transform features by scaling each feature to a given range.

```python
minmax_skl       = MinMaxScaler()
some_data_normed = minmax_skl.fit_transform(some_data)


fig, ax = subplots(nrows=1, ncols=2)

scaling = amax(some_data)*1.1

ax[0].scatter(some_data[:, 0], some_data[:, 1])
ax[0].set_xlim(-0.2, scaling)
ax[0].set_ylim(-0.2, scaling)
ax[0].set_xlabel('Some data')

ax[1].scatter(some_data_normed[:, 0], some_data_normed[:, 1], c='r')
ax[1].set_xlim(-0.2, scaling)
ax[1].set_ylim(-0.2, scaling);
ax[1].set_xlabel('MinMax Normalised data');
```

```python
minmax_skl.transform([[2.3, 1.7]])
```
